```python
# PROBLEM 202



def isprime(num):
    upto = num**(1/2)    
    if num==1:
        return False
    for i in range(2, round(upto)+1):
        if num % i == 0:
            return False
    return True

import math
def primeFactors(n): 
    ret = []
    # Print the number of two's that divide n 
    while n % 2 == 0: 
        ret.append(2), 
        n = n / 2
          
    # n must be odd at this point 
    # so a skip of 2 ( i = i + 2) can be used 
    for i in range(3, int(math.sqrt(n))+1, 2): 
          
        # while i divides n, print i ad divide n 
        while n % i == 0: 
            ret.append(i), 
            n = n / i 
              
    # Condition if n is a prime 
    # number greater than 2 
    if n > 2: 
        ret.append(n)
        
    return ret

print(primeFactors(12017639147))
print(primeFactors(1000001))
```
